# Division de datos en DatosEntrenamiento y Datos Prueba y entrenamiento del modelo

```python
"""Y ahora queda la parte en la que dividimos los datos que hemos obtenido.

Este dataset lo dividimos en dos grupos de datos, un dato para las pruebas pruebas finales 
y otra parte para datos de entrenamiento que utilizaremos en la siguiente fase para
entrenar el modelo de machine learning"""

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X=casas[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms',
       'Avg. Area Number of Bedrooms', 'Area Population', 'Price']]
y=casas['Price']
```

### Dividimos los datos

```python
#'Usamos el 70% de entrenamiento y 30% prueba'
#X_train   Nuevo data set con el 70% por cien del original. Los datos son aleatorios

X_train, X_test,y_train, y_test=train_test_split(X,y, test_size=0.3, random_state=42) 
```

Entrenar el modelo: Relizar la caja Modelo de entrenamiento

```python
lrm=LinearRegression()
lrm.fit(X_train,y_train)
```

# Crear el modelo de prueba: RElizar la caja Datos de pruebas

```python
from sklearn import metrics
```

```python
#creamos una variable que vamos a llamar predicciones que va a tener las predicciones de los precios
#de las casas para los datos de pruebas

predicciones=lrm.predict(X_test)
```

```python
#'Genera nuestras predicciones al pasar los datos de pruebas X_test. sera un array'
predicciones
```

```python
#Para ver en un grafico las diferencias entre las predicciones y valores reales (estan en Y_test)podemos usar un grafico de tipo scater. /
#Cuanto mas cercanos los puntos a la diagonal mejor sera
plt.scatter(y_test, predicciones)
plt.show()
```

```python
"""#Otra manera sería hacer un histograma
#donde le pasaríamos también los precios reales y se los restaríamos de las predicciones.
#Así calcularíamos nuestros valores
#pero a través de un histograma y realmente este valor sería el error.
#El valor residual entre el valor real es decir la diferencia entre el valor real y la predicción.
#Entonces vemos la distribución de los errores.
#Vemos que se acerca una distribución normal.
#Esto también significa que el modelo que hemos elegido es el correcto para abordar una estimación del
#precio mediante linear regression
#regresión lineal."""
sns.displot(y_test - predicciones)
plt.show()
```

```python
#Histograma
residuo=y_test - predicciones
sns.histplot(residuo, kde=True, bins=30)
plt.show()
```

### Metrica 1: MAE(Mean absolute error)=Media del valor absoluto de los errores

Nuestra intención es que cuanto menor sea este valor mejor es el modelo 

```python
metrics.mean_absolute_error(y_test, predicciones)
```

### Metrica 2: MSE (la media de los errores al cuadrado)

```python
metrics.mean_squared_error(y_test, predicciones)
```

### Metrica 3:  RMSE que sería la raíz cuadrada de la media de los errores al cuadrado 

```python
np.sqrt(metrics.mean_squared_error(y_test, predicciones))
```

# K-Medias
El algoritmo de k medias el algoritmo de k medias es un algoritmo de aprendizaje no supervisado
para resolver el problema de la clusterización.

Tiene como objetivo la partición de un conjunto de n observaciones en k grupos en el que cada observación
pertenece al grupo cuyo valor medio es más cercano.

Es un método utilizado en minería de datos

```python
import pandas as pd
import numpy as np
import matplotlib as plt
import matplotlib.pyplot as plt
import seaborn as sns
```

Vamos a importar la función make blobs lo que genera son puntos ejemplos aleatorios
para poder hacer unas pruebas sobre el algoritmo de las k-medias.

Lo que intenta es realmente organizar una serie de puntos en una serie de grupos.

```python
from sklearn.datasets import make_blobs
```

```python
# El num de puntos sera 200 (n_samples), num_caracteristicas (n_features) seran 2 (la x e y para poder representarlo), y el num de centros (centers), es decir cuantos grupos queremos hacer
datos=make_blobs(n_samples=200,n_features=2, centers=4)
# Aqui estan los datos
datos[0]
```

```python
#Aqui estan los grupos
datos[1]
```

```python
# Visualizar
plt.scatter(datos[0][:,0], datos[0][:,1])
plt.show()
```

Generamos algoritmo de las K-medias

```python
from sklearn.cluster import KMeans
```

```python
modelo=KMeans(n_clusters=4)
```

```python
modelo.fit(datos[0])
```

```python
#Aqui tenemos el punto central de cada uno de los grupos
modelo.cluster_centers_
```

```python
# Esta seria nuestra agrupacion
modelo.labels_
```

```python
# Esta la agrupacion que genero make_blobs
datos[1]
```

```python
#Generamos una fila con 2 columnas y en cada una un gráfico
fig, (ax1,ax2)=plt.subplots(1,2,figsize=(12,4))

#Grafico1
ax1.scatter(datos[0][:,0], datos[0][:,1],c=modelo.labels_)
ax1.set_title('Algoritmo de K-medias')

#Grafico2
ax2.scatter(datos[0][:,0], datos[0][:,1], c=datos[1])
ax2.set_title('Datos originales')

plt.show()
```

Hemos sido capaces de agruparlos en la misma forma que los originales
